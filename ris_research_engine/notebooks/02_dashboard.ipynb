# RIS Auto-Research Engine - Interactive Dashboard

5-tab interface for configuring, running, and analyzing experiments

## Initialize Dashboard

Load the interactive dashboard with all 5 tabs.

```python
from ris_research_engine.ui.jupyter_dashboard import RISDashboard

dashboard = RISDashboard(db_path="outputs/experiments/results.db")
dashboard.display()
```

## How to Use

- **Tab 1 (Configure)**: Set system parameters, choose probe/model, add to queue
- **Tab 2 (Run)**: Execute experiments, monitor progress
- **Tab 3 (Results)**: View metrics, training curves, compare experiments
- **Tab 4 (Analysis)**: Generate comparison plots, analyze campaigns
- **Tab 5 (Database)**: Browse all experiments, filter, export

### Quick Start Guide

1. Go to **Tab 1 (Configure)**:
   - Set M=4, K=16, N=16 for a quick test
   - Select probe type (e.g., random_uniform)
   - Select model type (e.g., mlp)
   - Set epochs to 10 for quick testing
   - Click "Add to Queue"

2. Go to **Tab 2 (Run)**:
   - Click "Run Queue" to execute
   - Watch progress in real-time

3. Go to **Tab 3 (Results)**:
   - Select your experiment from dropdown
   - View metrics and training curves

4. Go to **Tab 4 (Analysis)**:
   - Generate comparison plots

5. Go to **Tab 5 (Database)**:
   - Browse all experiments
   - Export to CSV/JSON
