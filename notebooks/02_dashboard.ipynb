## Launch Interactive Dashboard

The dashboard provides 5 tabs:

2. **Run**: Execute experiments and view progress
3. **Results**: Browse recent experiment results
4. **Analysis**: Compare probes, models, and analyze sparsity
5. **Database**: View and export all experiments

## Usage Instructions

### Configure Tab
- Set N (RIS elements), K (codebook size), M (sensing budget)
- Choose probe type and model architecture
- Adjust training parameters (epochs, learning rate, batch size)

### Run Tab
- Click "Run Experiment" to start
- Monitor progress and view results when complete

### Results Tab
- Click "Refresh Results" to see recent experiments
- View metrics and training information

### Analysis Tab
- Select analysis type (Probe/Model Comparison, Sparsity)
- Click "Run Analysis" to generate plots

### Database Tab
- Filter by campaign name
- Load and view all experiments
- Export to CSV
